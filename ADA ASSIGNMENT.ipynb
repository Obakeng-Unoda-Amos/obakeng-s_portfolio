b) Load the data (prices.csv) into a data frame and separate the predictor variable and the target variables into different files.

```python
#Read csv file
df = pd.read_csv('Downloads/ADA Assignment/Price.csv')

# separate the predictor variable and the target variables into different files
X = df.iloc[:, :-1].values
y = df.iloc[:,1].values 
```

c) Determine the correlation coefficient between the area and the price of houses and comment on the result.

```python
#using the pandas 'corr' function
#df.corr()
df.corr(method = 'pearson')
```

d) Using an 80/20 ratio, split the dataset into training set and test set

```python
# splitting the dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

e) Fit the linear regression algorithm into the training dataset

```python
# fitting the regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train) #produces the linear eqn for the data

#other method
#Fit the linear regression algorithm into the training dataset
###model = linear_model.LinearRegression()
#model = LinearRegression()
#model.fit(x_train, y_train)
```

f) Use the model to predict the prices of houses using the test set

```python
y_pred = regressor.predict(X_test) 
y_pred
 
y_test
```

g) On the same visualization, compare the scatter plot of the training dataset (X_train, y_train) and a line graph of X_train and the predicted price from the training set.

```python
# visualizing the results
# plot for the TRAIN
  
plt.scatter(X_train, y_train, color='red') # plotting the observation line
plt.plot(X_train, regressor.predict(X_train), color='blue') # plotting the regression line
plt.title("Area(sqm) vs Price($) (Training set)") # title of graph
  
plt.xlabel("Price($)") # name of x-axis
plt.ylabel("Area(sqm)") # name of y-axis
plt.show()
```

h) On the same visualization, compare the scatter plot of the test set (X_test, y_test) and a line graph of X_train and the predicted price from the training set.

```python
# plot for the TEST
  
plt.scatter(X_test, y_test, color='red') 
plt.plot(X_train, regressor.predict(X_train), color='blue') # plotting the regression line
plt.title("Area(sqm) vs Price($) (Testing set)")
  
plt.xlabel("Price($)") 
plt.ylabel("Area(sqm)") 
plt.show() 
```

i) Print out the coefficient of determination R2 

```python
# method(1)
#from sklearn.metrics import r2_score

# model evaluation for training set
#R_square = r2_score(y_train,y_train_predict)

# model evaluation for testing set
#R_square = r2_score(y_test,y_pred)
#print("coefficient_of_determination", R_square)

#-------------------------------------------------------------------------#

# method(2)
# training set
#regressor.score(X_train, y_train)

# testing set
#regressor.score(X_test, y_test)

#-------------------------------------------------------------------------#

# method(3)
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# model evaluation for training set
y_train_predict = regressor.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = r2_score(y_train, y_train_predict)

print("model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = regressor.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2 = r2_score(y_test, y_test_predict)

print("model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

### o) The model that we developed could suffer from overfitting. What is your understanding of the term overfitting?

Overfitting occurs when the model learns the detail and noise in the training data and tries to fit each data point on the curve to the extent that it negatively impacts the performance of the model on new data.

### p) Provide a brief explanation of any three (3) factors that lead to model overfitting and suggest anyone that could have affected our model.

1.Data used for training is not cleaned and contains noise.When data is not cleansed it may cause alot of problems to our model, this causes the model to capture the noise in the training data and fail to generalize the model's learning.

2.The size of the training dataset used is not enough.With small data the model tends to learn the training data too well, this then leads to overfitting.

3.The model has a high variance. A model with high variance may represent the data set accurately but could lead to overfitting to noisy or otherwise unrepresentative training data.
